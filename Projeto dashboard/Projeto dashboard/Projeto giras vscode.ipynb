# Projeto: Dashboard Interativo com Pandas — Lifestyle Weight Tracker


Este trabalho tem como objetivo analisar um conjunto de dados de hábitos de vida e entender quais fatores estão mais ligados à variação de peso (Weight_Change)

A ideia principal é:
1. carregar os dados

## 2) Verificar dados faltando

Aqui vamos identificar colunas com valores nulos (vazios).

Por que isso é importante?
- valores faltantes podem atrapalhar médias, correlações e gráficos
- antes de analisar, precisamos decidir como tratar esses dados
- essa decisão precisa ser justificada no trabalho

```python
faltando = df.isnull().sum()
faltando[faltando > 0]
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)

quality_report = pd.DataFrame({
    "missing_qtd": missing,
    "missing_%": missing_pct
})

print("Linhas duplicadas: ", df.duplicated().sum())
print("Não há linhas duplicadas")

display(quality_report[quality_report["missing_qtd"] > 0])

display(df.describe(include="all").T)
```

- Existem 2750 valores de Calories_Consumed (calorias consumidas) que estão negativas
- E os valores negativos/nulos de Workout_Type, são os valores das pessoas que não praticam atividades físicas
- Logo precisamos trocar apenas os valores de calorias que estão negativas, e não precisamos auterar os valores de Workout_Type

## 3) Tratamento dos dados

Tratamentos escolhidos (simples e eficientes):
- Calories_Consumed: valores vazios e negativos serão preenchidos com a mediana da coluna

Justificativa:
- é um método fácil de explicar
- mantém todas as linhas do dataset
- evita perder dados desnecessariamente
- Valores negativos foram substituídos pela mediana, pois ela representa melhor um valor e esse valor não é afetada por valores muito altos ou muito baixos (valores extremos)

```python
df.isnull().sum().sum() # total de nulos antes do tratamento
```

```python
df["Calories_Consumed"] = df["Calories_Consumed"].fillna(df["Calories_Consumed"].median())

df.isnull().sum().sum()  # total de nulos depois do tratamento, ficam apenas os valores nulos de Workout_Type, que é uma variável categórica.
```

Foram indentificado valores negativos que causam inconsistencia, por exemplo em "Calories_Consumed"

## 1. Corrigindo valores negativos em Protein_g, Carbs_g e Fat_g (14 registros cada)
Foram tradadas calorias negativas, mas faltou as colunas de macronutrientes, que também podem ter o mesmo problema. Valores negativos de proteína e carboidrato são impossíveis e podem precisar de tratamento igual ao aplicado em Calories_Consumed

```python
(df[['Protein_g', 'Carbs_g', 'Fat_g']] < 0).sum()
```

```python
df = df[(df['Protein_g'] >= 0) & 
        (df['Carbs_g'] >= 0) & 
        (df['Fat_g'] >= 0)]


cols = ['Protein_g', 'Carbs_g', 'Fat_g']

for col in cols:
    df.loc[df[col] < 0, col] = None


df[cols] = df[cols].fillna(df[cols].median())
```

```python
(df[['Protein_g', 'Carbs_g', 'Fat_g']] < 0).sum()
```

Foi escolhida a substituição pela mediana em todos os casos pois ela preserva a distribuição em casos menos sensiveis
